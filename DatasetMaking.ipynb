```
import numpy as np
import pandas as pd
```

```
date = []
for i in range(1 , 31):
  dates = "{da}/11/2020".format(da = i)
  date.append(dates)
```

```
dataset = pd.read_csv('NEWDATASET1.csv')
category = dataset.CATEGORY.unique()
```

```
dates = []
cat = []
for i in range(0 , 30):
  for j in range(0 , 18):
    dates.append(date[i])
    cat.append(category[j])
```

```
ds = pd.DataFrame()
```

```
ds.insert(0 , column="Date" , value=dates)
ds.insert(1 , column="Category" , value=cat)
```
